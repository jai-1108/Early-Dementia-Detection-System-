```python
from stable_baselines3 import DQN
import numpy as np

# Load trained DQN model
model = DQN.load("/content/drive/MyDrive/fyp/dqn_cognitive_task.zip")

cognitive_state_cols = X_test.columns[:3]
# Ensure selected columns exist and are numeric
for col in cognitive_state_cols:
    if col not in test_data.columns:
        raise ValueError(f"Column '{col}' is missing in the dataset!")
    if not np.issubdtype(test_data[col].dtype, np.number):  # Check if dtype is numeric
        try:
            test_data[col] = pd.to_numeric(test_data[col])  # Attempt conversion
        except ValueError:
            raise ValueError(f"Column '{col}' contains non-numeric values and cannot be converted.")

# Select 10 patients for testing
test_samples = test_data.sample(n=10, random_state=42).reset_index(drop=True)

# Task Mapping
task_mapping = {0: "Memory Task", 1: "Attention Task", 2: "Problem-Solving Task"}

# Run predictions for 10 patients
for i, row in test_samples.iterrows():
    sample = row[cognitive_state_cols].to_numpy().reshape(1, -1).astype(np.float32)  # Reshape and ensure float32 type
    action, _ = model.predict(sample)  # Predict task recommendation
    print(f"Patient {i+1} (ID: {row['patient_id']}): Recommended Task -> {task_mapping[int(action)]}")
```
